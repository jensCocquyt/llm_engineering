```python
# imports

from openai import OpenAI
from IPython.display import Markdown, update_display
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
openai = OpenAI()
OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
```

```python
# set up environment
system = """"
You are a experienced assistant that can answer questions about the code in any notebook. 
Always reference some documentation or other resources to support your answer.
You are answering to a student who is learning about LLMs.
Keep it simple and concise, never make assumptions.
Elaborate with examples when possible and appropriate.
If it is not a technical question, say so and don't answer.
"""
```

```python
# here is the question; type over this to ask something new



question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""

another_question = """
def ask(question, client, model):
    stream = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": system},
            {"role": "user", "content": question}
        ],
        stream=True
    )
    response = ""
    display_handle = display(Markdown(""), display_id=True)
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        update_display(Markdown(response), display_id=display_handle.display_id)
    return response
"""

non_technical_question = """
What is the capital of France?
"""

def ask(question, client, model):
    stream = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": system},
            {"role": "user", "content": question}
        ],
        stream=True
    )
    response = ""
    display_handle = display(Markdown(""), display_id=True)
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        update_display(Markdown(response), display_id=display_handle.display_id)
    return response
```

```python
# Get gpt-4o-mini to answer, with streaming

ask(question, openai, MODEL_GPT)
ask(another_question, openai, MODEL_GPT)
ask(non_technical_question, openai, MODEL_GPT)
```

```python
# Get Llama 3.2 to answer
ask(question, ollama,MODEL_LLAMA)
```
